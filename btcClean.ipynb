### Data Cleaning Bitcoin

#### Volume

```python
# import dataset
df = pd.read_csv("assets/crypto-markets.csv", parse_dates=["date"])
```

```python
btc = df[df['slug']=='bitcoin']
```

```python
btc = df[df['slug']=='bitcoin']
btc_vol = btc
bvol = btc_vol.groupby('date').sum()
```

```python
btc_vol[(btc_vol['date'] >= '2017-05-15') & (btc_vol['date'] <= '2018-11-29')]
```

```python
bvol[bvol.index >= '2017-05-15'].volume.head(10)
```

```python
import datetime

start_date = datetime.date(2017, 5, 15)
end_date = datetime.date(2018, 11, 29)
delta = datetime.timedelta(days=1)

while (start_date+(9*delta) <= end_date):
    coin_vol = btc_vol[(btc_vol['date']>=np.datetime64(start_date-(10*delta))) & (btc_vol['date']<=np.datetime64(start_date+(10*delta)))].sum()['volume']
    coin_vol = coin_vol - btc_vol[(btc_vol['date']==np.datetime64(start_date))].volume
    coin_vol_mean = coin_vol/20
    btc_vol.at[(btc_vol['date']==np.datetime64(start_date)),'volume'] = coin_vol_mean
    start_date = start_date+delta
```

```python
btc_vol[(btc_vol['date']>='2017-05-15') & (btc_vol['date']<='2018-11-29')]
```

```python
bvol2 = btc_vol.groupby('date').sum()
```

```python
bvol2[bvol2.index >= '2017-05-15'].volume.head(10)
```

```python
fig = px.scatter(bvol2, x = bvol2.index, y = bvol2['volume'],
                  height = 400, width = 800,color_discrete_sequence=['#33246f'])
fig.update_layout(xaxis_title = 'Date', yaxis_title='Volume')

fig.show()
```

```python
bigCounter = 0
smallCounter = 0
for i in range(len(btc_close)):
    if btc_close.loc[i, 'close'] > btc_close.loc[i,'high']:
        bigCounter = bigCounter +1
    elif btc_close.loc[i, 'close'] < btc_close.loc[i,'low']:
        smallCounter = smallCounter +1
print("Number of close > high: ", bigCounter)
print("Number of close < low: ", smallCounter)
```

```python
for i in range(len(btc_close)):
    if btc_close.loc[i, 'close'] > btc_close.loc[i,'high']:
        btc_close.at[btc_close.index ==i, 'close'] = (btc_close[btc_close.index ==i+1].open + btc_close[btc_close.index ==i].low) /2
    elif btc_close.loc[i, 'close'] < btc_close.loc[i,'low']:
        btc_close.at[btc_close.index ==i, 'close'] = (btc_close[btc_close.index ==i].high + btc_close[btc_close.index ==i].low) /2
```
